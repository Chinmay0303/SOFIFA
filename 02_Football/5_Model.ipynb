```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_columns',None)
```

```python
fifa=pd.read_csv('Clean_Data/SoFifa_data.csv',index_col=0)
fifa
```

```python
fifa.drop(columns='ID',inplace=True)
fifa
```

```python
## Numeric_features
Numeric_features=[feature for feature in fifa.columns if fifa[feature].dtype!='object']
```

```python
Numeric_features,len(Numeric_features)
```

```python
## finding correlation 

corr=fifa[Numeric_features].corr()
```

```python
corr_with_target=corr['Value_mill(€)']
```

```python
corr_with_target=corr_with_target.sort_values(ascending=False)
```

```python
corr_with_target
```

```python
corr[(corr['Value_mill(€)']>=0.2) | (corr['Value_mill(€)']<=-0.2)]['Value_mill(€)']
```

```python
Selected_features=corr[(corr['Value_mill(€)']>=0.2) | (corr['Value_mill(€)']<=-0.2)].index
```

```python
data=fifa[Selected_features]
data
```

```python
import statsmodels.api as sm
```

```python
X=data.drop(columns='Value_mill(€)')
X
```

```python
y=data['Value_mill(€)']
y
```

```python
X_intercept=sm.add_constant(X)
X_intercept
```

```python
model=sm.OLS(y,X_intercept).fit()
y_pred=model.predict(X_intercept)
model.summary()
```

```python
fifa['Predicted Value']=y_pred
fifa['Difference']=fifa['Predicted Value']-fifa['Value_mill(€)']
fifa
```

```python
fifa[['Name','Value_mill(€)','Predicted Value','Difference']]
```

```python
fifa[fifa['Difference']<-8][['Name','Value_mill(€)','Predicted Value','Difference']]
```
