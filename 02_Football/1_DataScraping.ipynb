```python
row=table.find('tr',class_="persist-header")
Columns=[]
prop=row.find_all('th')
for col in prop:
    if col.text=='':
        continue
    else:    
        Columns.append(col.text)
Columns,len(Columns)  
```

```python
def genearateTable(Rows,table):
    body=table.find('tbody')
    trows=body.find_all('tr')
    for trow in trows:
            col=[]
            tds=trow.find_all('td')
            a=tds[1].find('a')
            col.append(a.text)
            for td in tds[2:-1]:
                if (td.text!=''):
#                     print(td.text)
                    col.append(td.text)
                else:
    #                 print(td.text)
                    col.append(None)
            Rows.append(col)
```
