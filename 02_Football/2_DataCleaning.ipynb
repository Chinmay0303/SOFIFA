## Value in millions

```python
dataset['Value']
```

```python
# Removing '€'
data=dataset.copy()
data['Value']=data['Value'].str.replace('€','')
data.head(1)
```

```python
# if value contains M-> Value_mill(€)=Value
# if value contains K-> Value_mill(€)=Value/1000
# if value-0 -> 0


# Function to convert the values
def convert_value(value):
    if 'M' in value:
        return float(value[:-1])  # Remove 'M' and convert to float
    elif 'K' in value:
        return float(value[:-1]) / 1000  # Remove 'K', convert to float, and divide by 1000
    else:
        return float(value)

# Apply the function to the 'Value' column and create 'Value_mill(€)'
data['Value_mill(€)'] = data['Value'].apply(convert_value)

dataset=data.drop(columns='Value')
dataset
```

## Wages in Thousands

```python
dataset['Wage']
```

```python
## removing (€) from wages

data=dataset.copy()

data['Wage']=data['Wage'].str.replace('€','')

data.head(1)
```

```python
# if value contains K-> Wage_thous(€)=Wage
# if value does not contain K-> Wage_thous(€)=Wage/1000  (includes wage=0)

def convert_wage(wage):
    if 'K' in wage:
        return float(wage[:-1])
    else:
        return float(wage)/1000

data['Wage_thous(€)']=data['Wage'].apply(convert_wage)

dataset=data.drop(columns='Wage')
dataset
```

## Release clause in millions

```python
dataset['Release clause']
```

```python
# €42.8M = 42.8
# €394K = 0.394
# €0 = 0
```

```python
## removing (€) from release clause

data=dataset.copy()

data['Release clause']=data['Release clause'].str.replace('€','')

data.head(1)
```

```python
# if value contains M-> ReleaseClause_mill(€)=RC
# if value contains K-> ReleaseClause_mill(€)=RC/1000
# else same

def convert_rc(RC):
    if 'M' in RC:
        return float(RC[:-1])
    elif 'K' in RC:
        return float(RC[:-1])/1000
    else:
        return float(RC)

data['ReleaseClause_mill(€)']=data['Release clause'].apply(convert_rc)

dataset=data.drop(columns='Release clause')
```

## Feature Datatype 

```python
data.info()
```

```python
## Some features should be int but they are object because they contain '+' or '-'like Overall Rating, Potential, etc.

for feature in data.columns:
    if data[feature].dtype=='object':
        print(feature,data[feature].unique())
```

```python
data[data['Acceleration']=='92+7']
```

```python
Conv=[]
for feature in data.columns:
    # Check if the column's data type is 'object'
    if data[feature].dtype == 'object':
        # Check if any of the values in the column contain '+'
        if data[feature].str.contains('\+').any():
#             print(i,feature)
            Conv.append(feature)

            
```

```python
Conv[:-2],len(Conv[:-2])
```

```python
## ALl features are present in Conv which are to be converted from object to int
## FK Accuracy was left as it only contained '-' not '+'

## I have to remove the - or + sign in the features
## I am making a choice to not do the operation(- or +), instead I will keep the value before the sign only

dataset=data.copy()

def removeSign(feature):
    if '-' in feature:
        return int(feature.split('-')[0])
    elif '+' in feature:
        return int(feature.split('+')[0])
    else:
        return int(feature)

for feature in Conv[:-2]:
    dataset[feature]=dataset[feature].apply(removeSign)

## FK Accuracy was left as it only contained '-'
dataset['FK Accuracy']=dataset['FK Accuracy'].apply(removeSign)
```

```python
dataset['FK Accuracy'].unique()
```

```python
## Converting the datatype of Height and weight to integer

dataset['Height_cm']=dataset['Height_cm'].astype(int)
dataset['Weight_kg']=dataset['Weight_kg'].astype(int)
```
